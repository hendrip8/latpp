```python
df = df.drop(columns=['Sunshine','Evaporation','Cloud3pm','Cloud9am','Location','RISK_MM','Date'],axis=1)
df.shape
```

```python
#menghilangkan missing di raintommorow australia
df=df.dropna(subset=['RainTomorrow'])
df.shape
```

```python
#imputasi modus
from scipy.stats import mode
df['RainTomorrow'].fillna((df.RainTomorrow.mode())[0], inplace=True)
df['WindGustDir'].fillna((df.WindGustDir.mode())[0], inplace=True)
df['WindDir9am'].fillna((df.WindDir9am.mode())[0], inplace=True)
df['WindDir3pm'].fillna((df.WindDir3pm.mode())[0], inplace=True)
df['RainToday'].fillna((df.RainToday.mode())[0], inplace=True)
print(df.shape)
df.head(n=10)
```

```python
#imputasi mean
df['MinTemp'] = df.groupby('RainTomorrow')['MinTemp'].transform(lambda x: x.fillna(x.mean()))
df['MaxTemp'] = df.groupby('RainTomorrow')['MaxTemp'].transform(lambda x: x.fillna(x.mean()))
df['Rainfall'] = df.groupby('RainTomorrow')['Rainfall'].transform(lambda x: x.fillna(x.mean()))
df['WindGustSpeed'] = df.groupby('RainTomorrow')['WindGustSpeed'].transform(lambda x: x.fillna(x.mean()))
df['WindSpeed9am'] = df.groupby('RainTomorrow')['WindSpeed9am'].transform(lambda x: x.fillna(x.mean()))
df['WindSpeed3pm'] = df.groupby('RainTomorrow')['WindSpeed3pm'].transform(lambda x: x.fillna(x.mean()))
df['Humidity9am'] = df.groupby('RainTomorrow')['Humidity9am'].transform(lambda x: x.fillna(x.mean()))
df['Humidity3pm'] = df.groupby('RainTomorrow')['Humidity3pm'].transform(lambda x: x.fillna(x.mean()))
df['Pressure9am'] = df.groupby('RainTomorrow')['Pressure9am'].transform(lambda x: x.fillna(x.mean()))
df['Pressure3pm'] = df.groupby('RainTomorrow')['Pressure3pm'].transform(lambda x: x.fillna(x.mean()))
df['Temp9am'] = df.groupby('RainTomorrow')['Temp9am'].transform(lambda x: x.fillna(x.mean()))
df['Temp3pm'] = df.groupby('RainTomorrow')['Temp3pm'].transform(lambda x: x.fillna(x.mean()))
print(df.shape)
df.head()
```

```python
#cek missing
np.sum(df.isnull())
```

```python
#mengatasi outlier
from scipy import stats
z = np.abs(stats.zscore(df._get_numeric_data()))
print(z)
df= df[(z < 3).all(axis=1)]
print(df.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
c=pd.DataFrame([])
c['Keterangan']=["Tidak Outlier","Outlier"]
c['Jumlah Observasi']=[135201,6992]
bx = sns.barplot(x="Keterangan", y="Jumlah Observasi",data=c)
plt.show()
```

```python
#Transformasi Variabel Kategorik
df['RainToday'].replace({'No': 0, 'Yes': 1},inplace = True)
df['RainTomorrow'].replace({'No': 0, 'Yes': 1},inplace = True)

from sklearn import preprocessing
le=preprocessing.LabelEncoder()
aa=le.fit(df.iloc[0:,3])
a=le.transform(df.iloc[0:,3])
bb=le.fit(df.iloc[0:,5])
b=le.transform(df.iloc[0:,5])
cc=le.fit(df.iloc[0:,6])
c=le.transform(df.iloc[0:,6])

df['WindGustDir']=a
df['WindDir9am']=b
df['WindDir3pm']=c
df.head()
```

```python
#Transformasi Variabel Count
from sklearn import preprocessing
aa = df.drop(columns=['RainTomorrow','RainToday','WindGustDir','WindDir9am','WindDir3pm'])
bb=df[['RainTomorrow','RainToday','WindGustDir','WindDir9am','WindDir3pm']]
scaler = preprocessing.MinMaxScaler()
scaler.fit(aa)
aa = pd.DataFrame(scaler.transform(aa), index=aa.index, columns=aa.columns)
df=pd.concat([aa,bb],axis=1)
df.head()
print(df.shape)
```

```python
categorical_columns = ['WindGustDir', 'WindDir3pm', 'WindDir9am']
for col in categorical_columns:
    print(np.unique(df[col]))
```

```python
#feature selection (mencari skor)
from sklearn.feature_selection import SelectKBest, chi2
X = df.drop(columns=['RainTomorrow'])
y = df[['RainTomorrow']]
selector = SelectKBest(chi2, k='all').fit(X,y)
X_new = selector.transform(X)
scores = selector.scores_
```

```python
g=pd.DataFrame([])
g['Variabel']=list(X)
g['Skor']=scores.astype(int)
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
g=g.sort_values('Skor')
gx = sns.barplot(x="Skor", y="Variabel", data=g)
plt.show()
```

```python
#feature selection (mengambil 10 variabel terpenting)
from sklearn.feature_selection import SelectKBest, chi2
X = df.drop(columns=['RainTomorrow'])
y = df[['RainTomorrow']]
selector = SelectKBest(chi2, k=10).fit(X,y)
X_new = selector.transform(X)
print(X.columns[selector.get_support(indices=True)])
```
